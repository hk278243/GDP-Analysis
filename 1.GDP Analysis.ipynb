## Analysing Arab World

```python
df['Country Name'] == 'Arab World'
```

```python
df[df['Country Name'] == 'Arab World']
```

```python
df_pr = df[df['Country Name'] == 'Arab World']
```

```python
df_pr.plot(kind = 'line', x = 'Year', y = 'Value', figsize = (13, 5), grid = True, legend = True, ylabel = 'GDP', xlabel = 'Years')
```

```python
data = df_pr.values

gdp_change = [0]

for i in range(1, len(data)):
    
    prev = data[i-1][3]
    cur = data[i][3]
    
    gdp_change.append(round(((cur - prev) / prev) * 100, 2))
```

```python
df_pr.assign(GDP = gdp_change)
```
